```python
path = kagglehub.dataset_download("masoudnickparvar/brain-tumor-mri-dataset")
dataset_path = os.path.join(path, "Training")

print("Dataset path:", dataset_path)
print("Classes:", os.listdir(dataset_path))
```

```python
model = YOLO("yolov8n-cls.pt")  # lightweight & fast
```

```python
import shutil

# Original dataset path (read-only)
original_dataset_path = dataset_path

# Create a writable path in /content/
writable_dataset_base = "/content/brain-tumor-dataset"
writable_training_path = os.path.join(writable_dataset_base, "Training")

# Copy the original read-only dataset to the writable location
# Check if the writable_training_path already exists to avoid re-copying
if not os.path.exists(writable_training_path):
    print(f"Copying dataset from {original_dataset_path} to {writable_training_path}...")
    shutil.copytree(original_dataset_path, writable_training_path)
    print("Dataset copied successfully.")
else:
    print(f"Dataset already exists at {writable_training_path}. Skipping copy.")

# Update the dataset_path to the new writable location
dataset_path = writable_training_path

model.train(
    data=dataset_path,   # folder with class subfolders
    epochs=10,
    imgsz=224,
    batch=32,
    device="cpu",        # change to 0 if GPU available
    project="BrainTumorYOLOv8",
    name="yolov8_cls"
)
```
